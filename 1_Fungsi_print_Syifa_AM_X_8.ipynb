#Paramater Fungsi Print()

```python
Tahun = 2025
Bulan = "Agustus"
Tanggal = 23

print(Tahun, Bulan, Tanggal, sep="-")
```

```python
nama_depan = "Syifa"
nama_belakang = "Aurin Maghfiroh"

print(nama_depan, end=" ")
print(nama_belakang)
```

```python
nama_siswa_1 = "Dinda"
nama_siswa_2 = "Aurin"
nama_siswa_3 = "Lisaan"
print(nama_siswa_3, nama_siswa_2, nama_siswa_3, nama_siswa_1, sep='/')
```

```python
nama_siswa_4 = "Azkal"
nama_siswa_5 = "Hasim"
nama_siswa_6 = "Chilmi"
nama_siswa_7 = "Nova"

print(nama_siswa_6, end="~")
print(nama_siswa_4)
print(nama_siswa_5, end="-")
print(nama_siswa_7)
print(nama_siswa_4, end="/")
print(nama_siswa_7, end="&")
print(nama_siswa_5)
```
