#Contoh aplikasi sederhana

```python
panjang= int(input("Masukkan panjang dari persegi panjang(cm): "))
lebar= int(input("Masukkan lebar dari persegi panjang(cm): "))

luas = panjang * lebar

print(f"\nLuas dari persegi panjang yaitu {luas} cm^2")
```

#Mencari volume bola

```python
jari_jari = int(input("masukkan jari jari dari volume bola(cm): "))

volume = 4/3 * 22/7 * jari_jari **3

print(f"\nVolume dari bola yaitu {volume} cm^3")
```
